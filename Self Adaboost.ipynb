```python

from sklearn.datasets import make_classification
import warnings 
warnings.filterwarnings('ignore')
```

```python
x, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python

from sklearn.ensemble import AdaBoostClassifier
```

```python
ada_clf=AdaBoostClassifier()
ada_clf
```

```python
ada_clf.fit(x_train,y_train)
ada_clf
```

```python
y_pred=ada_clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0, 1.5, 2.0],
    'algorithm': ['SAMME', 'SAMME.R']
}

grid_obj = GridSearchCV(estimator=ada_clf,param_grid=param_grid, cv=5, verbose = 3) 
grid_obj
#if njobs = -1, you will not gget all fitting information
```

```python
grid_obj.fit(x_train,y_train)
```

```python
grid_obj.best_params_
```

```python
model_made_after_tunning=grid_obj.best_estimator_
model_made_after_tunning
```

```python
y_pred_tuned=model_made_after_tunning.predict(x_test)
y_pred_tuned
```

```python
print(accuracy_score(y_test,y_pred_tuned))
print(classification_report(y_test,y_pred_tuned))
print(confusion_matrix(y_test,y_pred_tuned))
```
